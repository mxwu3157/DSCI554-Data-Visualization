```python
import pandas as pd
import json
```

```python
df = pd.read_csv('covid19_year_to_year.csv')
```

```python
df
```

```python
result = df.to_json(orient="records")
parsed = json.loads(result)
json_obj = json.dumps(parsed)  

with open("covid19_year_to_year.json", "w") as outfile:
    outfile.write(json_obj)
```

```python
df = pd.read_csv('../data/co2_emission_country_cont.csv')
df.head()
```

```python
result = df.to_json(orient="records")
parsed = json.loads(result)
json_obj = json.dumps(parsed)  

with open("../data/co2_emission_country_cont.json", "w") as outfile:
    outfile.write(json_obj)
```

```python
df2 = df[['country','emission']]

df2.to_csv('../data/co2_emission_country.csv', index=False)
result = df2.to_json(orient="records")
parsed = json.loads(result)
json_obj = json.dumps(parsed) 
with open("../data/co2_emission_country.json", "w") as outfile:
    outfile.write(json_obj)
df2
```
